**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here
x = input("Enter a number:")
f = 1
f = f*int(x)
for i in range(1,int(x)):
  f = f*(int(x)-i)
print(f, "is the factorial of", x)
```
